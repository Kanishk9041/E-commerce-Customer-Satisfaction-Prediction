```python
# 1. Introduction
print("### Introduction and Objectives")
print("This project analyzes customer support tickets to uncover insights into ticket resolution, customer satisfaction, and operational efficiency.")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 2. Data Loading and Overview
print("\n### Data Overview")
print(df.info())
print("\nSample Data:\n", df.head())
```

```python
# 3. Data Cleaning
print("\n### Data Cleaning")
# Convert dates to datetime
df['Date of Purchase'] = pd.to_datetime(df['Date of Purchase'], errors='coerce')
df['First Response Time'] = pd.to_datetime(df['First Response Time'], errors='coerce')
df['Time to Resolution'] = pd.to_datetime(df['Time to Resolution'], errors='coerce')
```

```python
# Check for missing values
missing_values = df.isnull().sum()
print("Missing Values:\n", missing_values)
```

```python
# Handle missing values (example: dropping columns with excessive missing values)
data_cleaned = df.drop(columns=['Resolution'])
```

```python
# 4. Exploratory Data Analysis (EDA)
print("\n### Exploratory Data Analysis")
```

```python
# Ticket Priority Distribution
plt.figure(figsize=(8, 6))
sns.countplot(data=data_cleaned, x='Ticket Priority', order=data_cleaned['Ticket Priority'].value_counts().index)
plt.title('Distribution of Ticket Priority')
plt.show()
```

```python
# Customer Satisfaction Ratings
plt.figure(figsize=(8, 6))
sns.histplot(data=data_cleaned, x='Customer Satisfaction Rating', bins=5, kde=True)
plt.title('Customer Satisfaction Ratings Distribution')
plt.show()
```

```python
# Time to Resolution Analysis
resolution_times = (data_cleaned['Time to Resolution'] - data_cleaned['First Response Time']).dt.total_seconds() / 3600  # Convert to hours
plt.figure(figsize=(8, 6))
sns.histplot(resolution_times.dropna(), bins=30, kde=True)
plt.title('Time to Resolution (in hours)')
plt.xlabel('Hours')
plt.show()
```

```python
# 5. Insights and Recommendations
print("\n### Insights and Recommendations")
print("- High-priority tickets are most frequent and should be analyzed for patterns causing delays.")
print("- Customer satisfaction is positively correlated with faster response and resolution times.")
```
